# openSUSE translations
openSUSE uses its own translation website, but it is also based on the Weblate API. Hence, it is possible to quickly adapt the previous code to collect all the translation of the software, that are numerous.

```python
# list of weblate apis and the associated desired project. when ALL : all the projects will be downloaded
wanted_os = ["ALL;openSUSE"]
collector.get_all_translations("https://l10n.opensuse.org/api/", "", wanted_os)
```

# Ubuntu Launchpad translations

**Note :** the dataset could be extended with previous versions of Ubuntu, that are also available on Launchpad.

```python
collector.get_ubuntu_translation("ubuntu-noble", "https://translations.launchpad.net/ubuntu/noble/+translations")
```
