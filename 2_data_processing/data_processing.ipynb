## Structure of the dataset
- A folder for training data : Open source operating system translations (except Ubuntu)
- A folder for test data : Ubuntu translations
- Each folder has one subfolder per language
- Each folder will have a text folder saving : 
    - the last time the dataset has been updated (using ../data_collection tools)
    - the last time the dataset has been pre-processed
